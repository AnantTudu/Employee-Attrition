# Data Analysis

```python
# Descriptive statistics for age
age_distribution = df['Age'].describe()

# Print age distribution
print("Distribution of Age:")
print(age_distribution)
```

```python
# Get the counts of attrition
attrition_counts = df['Attrition'].value_counts()

# Print the counts
print(attrition_counts)
```

```python
# Calculate attrition rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100

# Print attrition rate
print("Attrition Rate:")
print(attrition_rate)
```

```python
# Group data by monthly income and attrition status and get the counts
attrition_by_income = df.groupby(['Attrition'])['MonthlyIncome'].value_counts()

# Get the top 10 counts
top_10_counts = attrition_by_income.nlargest(10)

# Print the top 10 counts
print(top_10_counts)
```

```python
# Group data by marital status and attrition status and get the counts
attrition_by_marital_status = df.groupby(['Attrition'])['MaritalStatus'].value_counts()

# Get the top 10 counts
top_10_counts = attrition_by_marital_status.nlargest(10)

# Print the top 10 counts
print(top_10_counts)
```

```python
# Group data by department and attrition status and get the counts
attrition_by_department = df.groupby(['Attrition'])['Department'].value_counts()

# Get the top 10 counts
top_10_counts = attrition_by_department.nlargest(10)

# Print the top 10 counts
print(top_10_counts)

```

```python
# Group data by gender and attrition status and get the counts
attrition_by_gender = df.groupby(['Attrition', 'Gender']).size()

# Get the top 10 counts for each gender
top_10_male = attrition_by_gender['Yes', 'Male']
top_10_female = attrition_by_gender['Yes', 'Female']

# Print the top 10 counts for each gender
print("Attrition by gender:")
print("Male:", top_10_male)
print("Female:", top_10_female)
```

```python
# Group data by education field and attrition status and get the counts
attrition_by_education_field = df.groupby(['Attrition', 'EducationField']).size()

# Get the top 10 counts for each education field
top_10_counts_education_field = attrition_by_education_field['Yes'].nlargest(10)

# Print the top 10 counts for each education field
print("Top 10 counts of attrition by education field:")
print(top_10_counts_education_field)
```

# Data Visualization

```python
# Distribution analysis
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize attrition rate
plt.figure(figsize=(6, 6))
df['Attrition'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Attrition Rate')
plt.ylabel('')
plt.show()
```

```python
# Attrition by education field
plt.figure(figsize=(10, 6))
sns.countplot(x=df['EducationField'], hue=df['Attrition'])

plt.title('Attrition by Education Field')
plt.xlabel('Education Field')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.xticks(rotation=45)
plt.show()
```

```python
# Attrition by gender
plt.figure(figsize=(8, 6))
sns.countplot(x=df['Gender'], hue=df['Attrition'])
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
# Attrition by department
plt.figure(figsize=(8, 6))
sns.countplot(x=df['Department'], hue=df['Attrition'])
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
# Attrition by Marital Status
plt.figure(figsize=(8, 6))
sns.countplot(x=df['MaritalStatus'], hue=df['Attrition'])
plt.title('Attrition by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

```python
# Attrition by Monthly Income
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='MonthlyIncome', hue='Attrition', kde=True, bins=30)
plt.title('Attrition by Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.show()
```

# Insights

1. Out of 4300 employees, 695 have left the company,while 3605 did not leave , resulting in a 16.2% attrition rate.

2. 430 male and 265 female employees contributed to the attrition statistics.

3. Maximum attrition is done by employee of lower salary, as the employee salary increases attrition decreases.

4. Single employees exhibit the highest attrition rate compared to married and divorced employees. 
Married employees experience a lower attrition rate than single employees but a higher attrition rate than divorced employees. 
Divorced employees have the lowest attrition rate among the three marital status categories.

5. The Research & Development department exhibits the highest attrition rate compared to the Sales and Human Resources 
departments. Sales department employees experience a higher attrition rate than those in the Human Resources department.

6. Male employees exhibit a higher attrition rate compared to female employees across departments. 
This observation suggests that there may be gender-specific factors influencing attrition, 
such as career advancement opportunities, work-life balance, or organizational culture. 

7. Employees with educational backgrounds in Life Sciences and Medical fields exhibit higher attrition rates compared to 
those from Marketing, Technical Degree, Human Resources, and Other fields. This observation suggests that there may be 
specific challenges or opportunities within each educational background that contribute to attrition rates. Understanding 
these trends can help tailor retention strategies, such as career development programs or targeted support initiatives, to 
address the unique needs of employees from different educational backgrounds and mitigate attrition effectively.
